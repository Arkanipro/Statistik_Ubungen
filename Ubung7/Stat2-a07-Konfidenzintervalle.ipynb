### Task 2: Bootstrap Confidence Interval Estimation
- As previously, randomly extract a sample of $N$ data points from the data set `americans.csv`.
- Vary the size of the sample $N$ from 
$ N \in \{10, 50, 100\}$
 data points
- Calculate the bootstrap confidence intervals for the three estimates:
    - mean height (cm)
    - mean weight (kg)
    - count ("does the groceries") out of $N$ (how many of the $N$ do the groceries)  
- Use 1000 bootstrap samples (only calculate one bootstrap confidence interval for each feature)  
- You can use the numpy function [numpy.quantile](https://numpy.org/doc/stable/reference/generated/numpy.quantile.html) to calculate the quantiles.
- Use the percentile and pivot method to calculate the confidence intervals.  
- Store the two results ($3 \cdot 1000$ bootstrap estimates and $3 \cdot 2$ confidence intervals) in two separate pandas data frames
- Calculate the corresponding bootstrap confidence interval when sampling 1000 times $N$ samples with replacement from the full data set.
- Extract the first confidence interval from task 1
- Visualize the different confidence intervals (from the bootstrap sample, from the population and the first from task 1)=
- Print the numeric values of the different confidence intervals.






`````python
import pandas
import numpy
import plotly.graph_objects 
import scipy.stats
import plotly.express 

numpy.random.seed(21)
repeat=1000
N=[10,50,100]

americans=pandas.read_csv("americans.csv",delimiter=",")

def Konfidenzintervall_Normalverteilung(mean,var,n,z=1.96):
    x=z*(var/(numpy.sqrt(n)))
    return [mean-x,mean+x]
    
americans
`````

`````python
df=pandas.DataFrame()
for n in N:
    conf_int_height_unten,conf_int_height_oben,conf_int_weight_unten,conf_int_weight_oben=[],[],[],[]
    for i in range(repeat):
        sample=americans.sample(n=n)
        ret=Konfidenzintervall_Normalverteilung(sample['Height (cm)'].mean(),numpy.sqrt(sample['Height (cm)'].var()),n)
        conf_int_height_oben.append(ret[1])
        conf_int_height_unten.append(ret[0])
        ret=Konfidenzintervall_Normalverteilung(sample['Weight (kg)'].mean(),numpy.sqrt(sample['Weight (kg)'].var()),n)
        conf_int_weight_oben.append(ret[1])
        conf_int_weight_unten.append(ret[0])
        #norm_height=scipy.stats.norm(loc=sample['Height (cm)'].mean(),scale=numpy.sqrt(sample['Height (cm)'].var()))
        #norm_weight=scipy.stats.norm(loc=sample['Weight (kg)'].mean(),scale=numpy.sqrt(sample['Weight (kg)'].var()))
    df['untere Größe n='+str(n)]=conf_int_height_unten
    df['obere Größe n='+str(n)]=conf_int_height_oben
    df['unteres Gewicht n='+str(n)]=conf_int_weight_unten
    df['oberes Gewicht n='+str(n)]=conf_int_weight_oben
df      
`````
